# Processing on DF

```python
df["month"]=pd.to_datetime(df["date"]).dt.month
df["year"]=pd.to_datetime(df["date"]).dt.year
```

```python
modified_df=df[['year','month','iso_code', 'continent','location','total_cases','new_cases', 'total_deaths', 'new_deaths', 'people_vaccinated','people_fully_vaccinated','population']]
```

```python
grouped_df=modified_df.groupby(["year","month","continent","location","iso_code"]).sum()
```

```python
grouped_df.shape
```

```python
pd.set_option('display.max_rows', 500)
grouped_df.head(500)
```

```python
grouped_df.isnull().sum().sum()
```

```python
grouped_df["year"]=grouped_df.index.get_level_values(0)
grouped_df["month"]=grouped_df.index.get_level_values(1)
grouped_df["continent"]=grouped_df.index.get_level_values(2)
grouped_df["location"]=grouped_df.index.get_level_values(3)
grouped_df["iso_code"]=grouped_df.index.get_level_values(4)
```

```python
month={1:"january",2:"February",3:"March",4:"April",5:"May",6:"June",7:"July",8:"August",9:"Septemper",10:"October",
       11:"Novamber",12:"December"}
```

```python
grouped_df["month"]=grouped_df["month"].map(month)
```

```python
grouped_df.reset_index(drop=True,inplace=True)
```

## Building our Dash

```python
from logging import PlaceHolder
import dash
from dash import html
from dash import dcc
from dash.dependencies import Input,Output,State


app=dash.Dash(external_stylesheets=['https://codepen.io/chriddyp/pen/bWLwgP.css'])
app.layout = html.Div(
    children=[
    html.Div(
    children=[
    html.P("Covid-19 Summary",style={"font-size":30,"font-weight":"bold","color":"Black"}),
    
    
        
    dcc.Slider(id='my_slider',min=grouped_df['year'].min(),max=grouped_df['year'].max(),value=grouped_df['year'].min(),
    step=None,marks={str(years): str(years) for years in grouped_df["year"].unique()}),
        
    dcc.Dropdown(id='my_month',
        options=[{"label":k,"value":k} for k in lst],
        value='january'), dcc.Graph(id='my_graph',figure=fig1)],className='eight columns'),
    
    html.Br(),
    
    html.Div(children=[dcc.Dropdown(id="my_continent",options=[
        {"label":k,"value":k} for k in lst_continent]
        ,value="Asia",style={"backgroundColor": "Reds"}),
    html.Br(),
    html.P("Total Cases Confirmed",style={"font-size":18,"font-weight":"bold"}),
    html.P(f"+ {total_confirmed_cases}",style={"font-size":18},id="cases"),
    html.Br(),
    html.P("Total Deaths",style={"font-size":18,"font-weight":"bold"}),
    html.P(f"+ {total_deaths}",style={"font-size":18},id="deaths"),
    html.Hr(),
    dcc.Graph(id="bar_graph",figure=fig2)
                                                    
    ]
    ,className='four columns')])

@app.callback(
    Output(component_id='my_graph',component_property='figure'),
    Output(component_id="bar_graph",component_property="figure"),
    Output(component_id="cases",component_property="children"),
    Output(component_id="deaths",component_property="children" ),
    Input(component_id='my_slider',component_property='value'),
    Input(component_id='my_month',component_property='value'),
    Input(component_id="my_continent",component_property="value"),
    
)
#Four Cases of Connditions
def update_My_graph(input_value,my_month,my_continent):
    
        filtered_df=grouped_df[(grouped_df.month==my_month)&(grouped_df.year==input_value)]
        fig_1=px.choropleth(filtered_df, locations="iso_code",featureidkey="properties.iso_code",color='total_cases',
                     hover_name='continent',labels={"total_cases":"Total_Cases"},animation_frame="month",hover_data=['location','year','month','total_deaths','people_fully_vaccinated'] ,
                                  color_continuous_scale=px.colors.sequential.Reds).update_layout(margin={"r":0,"t":0,"l":0,"b":0}).update_geos(bgcolor="#1f77b4",framewidth=0)
        fig_2=px.bar(filtered_df, x="continent", y=["total_cases", "total_deaths"],height=330).update_layout(margin={"r":0,"t":0,"l":0,"b":0})
        num_cases=grouped_df[(grouped_df["year"]==input_value)&(grouped_df["month"]==my_month)&(grouped_df["continent"]==my_continent)].total_cases.sum()
        num_deaths=grouped_df[(grouped_df["year"]==input_value)&(grouped_df["month"]==my_month)&(grouped_df["continent"]==my_continent)].total_deaths.sum()
        
        return fig_1,fig_2,num_cases,num_deaths
        
if __name__ == '__main__':
    app.run_server()
```
